```python
#!pip install MLB-StatsAPI

import statsapi
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#pull the top 10 hitters by batting average for the 2024 season
leaders = statsapi.league_leader_data('battingAverage', season = 2024, statGroup = 'hitting', limit = 10)
```

```python
#loop through list of top 10 players, unpack data, and build a player dictionary
data = []
for row in leaders:
    rank, player, team, avg = row
    data.append({'Player': player, 'Team': team, 'AVG': float(avg)})
```

```python
#convert list into pandas dataframe
df = pd.DataFrame(data)
df = df.sort_values(by = 'AVG', ascending = False)

print(df)
```

```python
#plot a bar chart
plt.figure(figsize = (10, 6))
plt.barh(df['Player'], df['AVG'], color = 'skyblue')
plt.xlabel('Batting Average')
plt.title('Top 10 MLB Batting Averages (2024 Season)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
